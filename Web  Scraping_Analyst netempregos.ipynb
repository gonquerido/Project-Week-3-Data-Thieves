```python
# the main library you will need for webscraping is called Beautiful Soup

from bs4 import BeautifulSoup

# the second package we will need we already know it

import requests
```

```python
url = 'https://www.net-empregos.com/pesquisa-empregos.asp?chaves=&cidade=&categoria=0&zona=0&tipo=0'

response = requests.get(url)
response
```

```python
response.content
```

```python
soup = BeautifulSoup(response.content)
```

```python
print(soup.prettify())
```

```python
soup.title.string
```

```python
soup.p
```

```python
divisions = soup.find_all('div')
divisions
```

```python
jobs = soup.find_all('div', attrs = {'class' : "job-item job-item-destaque media"})
```

```python
job = jobs[0]
job
```

```python
for line in job.find_all('li'):
  print(line.text)
```

```python
for line in job.find_all('div'):
  print(line.text)
```
