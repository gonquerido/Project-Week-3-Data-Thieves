```python
from bs4 import BeautifulSoup

import requests
```

```python
url = ("https://www.paylab.com/pt/salaryinfo")

response = requests.get(url)
response
```

```python
response.content
```

```python
soup = BeautifulSoup(response.content, features = "lxml")
```

```python
type(soup)
```

```python
print(soup.prettify())
```
